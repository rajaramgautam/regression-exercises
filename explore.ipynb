### 4. Run the appropriate statistical tests where needed.

```python
# Statistical Tests for 4 Hypothesis:
from scipy import stats
def pearsonr(col1, col2):
    α = 0.05
    corr, p = stats.pearsonr(col1, col2)
    corr, p

    if p < α:
        print("We reject the null hypothesis")
    else:
        print("We fail to reject the null hypothesis")
    print(p)
```

```python
def spearmanr(col1, col2):   
    α = 0.05
    r, p_value = stats.spearmanr(col1, col2)
    r, p_value

    if p_value < α:
        print("We reject the null hypothesis")
    else:
        print("We fail to reject the null hypothesis")
    print(p_value)
```

```python
# Hypothesis 1
pearsonr(train.area, train.tax_value)
```

```python
spearmanr(train.area, train.tax_value)
```

```python
# Hypothesis 2
pearsonr(train.bathrooms, train.tax_value)
```

```python
spearmanr(train.bathrooms, train.tax_value)
```

```python
# Hypothesis 3
pearsonr(train.bedrooms, train.tax_value)
```

```python
spearmanr(train.bathrooms, train.tax_value)
```

```python
# Hypothesis 4
pearsonr(train.age, train.tax_value)
```

```python
spearmanr(train.age, train.tax_value)
```
